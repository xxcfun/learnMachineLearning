```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
cosy = np.cos(x)
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
```

```python
plt.scatter(x, siny)
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red")
```
